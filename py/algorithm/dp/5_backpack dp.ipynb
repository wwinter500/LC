```python
#package dp
#92 - backpack I
#dp[i][j] = if can fullfil j size when at most i items
def backPack(m, A):
    if not A:
        return 0

    n = len(A)
    dp = [[False for j in range(m + 1)] for i in range(n + 1)]
    for i in range(n + 1):
        dp[i][0] = True

    ans = 0
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if j >= A[i - 1]:
                dp[i][j] = dp[i - 1][j] or dp[i - 1][j - A[i - 1]]
            else:
                dp[i][j] = dp[i - 1][j]

    for j in range(m, -1, -1):
        if dp[n][j]:
            return j

    return 0
print(backPack(100, [16,16,3,5,16,16,16,16,16,16,16,16,16,16,16,16,16]))
```

```python
#666 - guess number II 
#dp[i][j] - minimun money to pay to finish the game
#dp[i][j] = max(dp[i][k - 1], dp[k + 1][j]) + k
import sys
def getMoneyAmount(n):
    dp = [[0 for j in range(n + 1)] for i in range(n + 1)]
    for l in range(2, n + 1):
        for s in range(1, n - l + 2):
            e = s + l - 1
            cost = sys.maxsize
            for k in range(s, e + 1):
                if k == e:
                    cost = min(cost, dp[s][k - 1] + k)
                elif k == s:
                    cost = min(cost, dp[k + 1][e] + k)
                else:
                    cost = min(cost, max(dp[s][k - 1], dp[k + 1][e]) + k)
                
            dp[s][e] = cost
    
    for i in range(n):
        print(dp[i])
    return dp[1][n]
print(getMoneyAmount(10))
```

```python
#backpack - III
#dp[i][j] - maximun values with j size when at most i items
def backPackIII(A, V, m):
    if not A or not V:
        return 0
    
    n = len(A)
    dp = [[0 for j in range(m + 1)] for i in range(n + 1)]
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if j >= A[i - 1]:
                times = j // A[i - 1] + 1
                for t in range(1, times):
                    dp[i][j] = max(dp[i - 1][j], dp[i - 1][j - A[i - 1] * t] + V[i - 1] * t)
            else:
                dp[i][j] = dp[i - 1][j]
     
    return max(dp[n])
print(backPackIII([2, 3, 5, 7], [1, 5, 2, 4], 10))
```

```python
#562 - backpack - IV 
#dp[i][j] - ways of combines with j size when at most i items
def backPackIV(nums, target):
    if not nums:
        return 0
    
    n = len(nums)
    dp = [[0 for j in range(target + 1)] for i in range(n + 1)]
    for i in range(n + 1):
        dp[i][0] = 1

    for i in range(1, n + 1):
        for j in range(1, target + 1):
            k = 0
            while k * nums[i - 1] <= j:
                dp[i][j] += dp[i - 1][j - nums[i - 1] * k]
                k += 1

    return dp[n][target]
print(backPackIV([2,3,6,7], 7))
```

```python
#backpack - II 
#dp[i][j] - maximum value with j size when at most i items
def backPackII(m, A, V):
    if not A or not V:
        return 0
    
    n = len(A)
    dp = [[0 for j in range(m + 1)] for i in range(n + 1)]
    
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if j >= A[i - 1]:
                dp[i][j] = max(dp[i - 1][j], dp[i - 1][j - A[i - 1]] + V[i - 1])
            else:
                dp[i][j] = dp[i - 1][j]
    
    return max(dp[n])
```
