```python
#ladder-2 index dp
def uniquePathsWithObstacles(obstacleGrid):
    if not obstacleGrid or not obstacleGrid[0]:
        return 0

    n,m = len(obstacleGrid), len(obstacleGrid[0])
    dp = [[0 for j in range(m)] for i in range(n)]
    for i in range(n):
        if obstacleGrid[i][0] == 1:
            break

        dp[i][0] = 1

    for j in range(m):
        if obstacleGrid[0][j] == 1:
            break

        dp[0][j] = 1

    for i in range(1, n):
        for j in range(1, m):
            if obstacleGrid[i][j] == 0:
                dp[i][j] = dp[i - 1][j] + dp[i][j - 1]  

    return dp[n - 1][m - 1]
```

```python
#515 - paint house
def minCost(costs):
    nhouse = len(costs)
    dp = [costs[0][0],costs[0][1],costs[0][2]]
    for i in range(1, nhouse):
        r,b,g = dp[0], dp[1], dp[2]
        dp[0] = min(b, g) + costs[i][0]
        dp[1] = min(r, g) + costs[i][1]
        dp[2] = min(r, b) + costs[i][2]

    return min(dp)
```

```python
def numDecodings(s):
    if not s:
        return 0
    
    n = len(s)
    dp = [0] * (n + 1)
    dp[0] = 1
    dp[1] = 1 if s[0] != '0' else 0
    for i in range(2, n + 1):
        val = int(s[i - 2:i])
        if s[i - 1] == '0':
            if val == 0 or val >= 30:
                return 0
            else:
                dp[i] = dp[i - 2]
        elif val > 26 or val < 10:
            dp[i] = dp[i - 1]
        else:
            dp[i] = dp[i - 1] + dp[i - 2]

    return dp[n]
```

```python
#397 - longest continues increasing subsequence
def longestIncreasingContinuousSubsequence(A):#LCS
        if not A:
            return 0
        
        n = len(A)
        lpre, rpre = 1,1
        ans = 1
        for i in range(1, n):
            if A[i] > A[i - 1]:
                lpre += 1
                ans = max(ans, lpre)
            else:
                lpre = 1
            
        for i in range(1, len(A)):
            if A[n - 1 - i] > A[n - i]:
                rpre += 1
                ans = max(ans, rpre)
            else:
                rpre = 1
        
        return ans
```

```python
#553 - bomb enemy
def maxKilledEnemies(grid):
    if not grid or not grid[0]:
        return 0
    
    n,m = len(grid), len(grid[0])
    up = [[0 for j in range(m)] for i in range(n)]
    down = [[0 for j in range(m)] for i in range(n)]
    left = [[0 for j in range(m)] for i in range(n)]
    right = [[0 for j in range(m)] for i in range(n)]
    
    for j in range(m):
        up[0][j] = 1 if grid[0][j] == 'E' else 0
        down[n - 1][j] = 1 if grid[n - 1][j] == 'E' else 0
    
    for i in range(n):
        left[i][0] = 1 if grid[i][0] == 'E' else 0
        right[i][m - 1] = 1 if grid[i][m - 1] == 'E' else 0
        
    for i in range(1, n):
        for j in range(m):
            if grid[i][j] == 'W':
                up[i][j] = 0
            elif grid[i][j] == '0':
                up[i][j] = up[i - 1][j]
            else:
                up[i][j] = up[i - 1][j] + 1
                
    for i in range(n - 2, -1, -1):
        for j in range(m):
            if grid[i][j] == 'W':
                down[i][j] = 0
            elif grid[i][j] == '0':
                down[i][j] = down[i + 1][j]
            else:
                down[i][j] = down[i + 1][j] + 1
                
    for j in range(1, m):
        for i in range(n):
            if grid[i][j] == 'W':
                left[i][j] = 0
            elif grid[i][j] == '0':
                left[i][j] = left[i][j - 1]
            else:
                left[i][j] = left[i][j - 1] + 1
    
    for j in range(m - 2, -1, -1):
        for i in range(n):
            if grid[i][j] == 'W':
                right[i][j] = 0
            elif grid[i][j] == '0':
                right[i][j] = right[i][j + 1]
            else:
                right[i][j] = right[i][j + 1] + 1
    ans = 0
    for i in range(n):
        for j in range(m):
            if grid[i][j] == '0':
                ans = max(ans, up[i][j] + down[i][j] + left[i][j] + right[i][j])
        
    return ans
print(maxKilledEnemies(["0E00", "E0WE", "0E00"]))
```

```python
#delete operation to find substring
def minDeleteOperation(word1, word2):
    n, m = len(word1), len(word2)
    f = [[0] * (m + 1) for _ in range(n + 1)]#dp array for longest matched subsequence
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            f[i][j] = max(f[i - 1][j], f[i][j - 1], f[i - 1][j - 1] + (word1[i - 1] == word2[j - 1]))
    
    return n + m - 2 * f[n][m]
minDeleteOperation("abcdeca", "acedcba")
```

```python
#Maximum Number of Points with Cost
def maxPoints(P):
    m, n = len(P), len(P[0])
    if m == 1: return max(P[0])
    if n == 1: return sum(sum(x) for x in P)

    def left(arr):
        lft = [arr[0]] + [0] * (n - 1)
        for i in range(1, n): lft[i] = max(lft[i - 1] - 1, arr[i])
        return lft

    def right(arr):
        rgt = [0] * (n - 1) + [arr[-1]]
        for i in range(n - 2, -1, -1): rgt[i] = max(rgt[i + 1] - 1, arr[i])
        return rgt

    pre = P[0]
    for i in range(m - 1):
        lft, rgt, cur = left(pre), right(pre), [0] * n
        for j in range(n):
            cur[j] = P[i + 1][j] + max(lft[j], rgt[j])
        pre = cur[:]

    return max(pre)
```
