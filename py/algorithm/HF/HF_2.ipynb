```python
#417 valid number: 
def isvalid(s):
    l, r = 0, len(s) - 1
    while l <= r and s[l] in ' \t':
        l += 1
    while l <= r and s[r] in ' \t':
        r -= 1
    if l > r:
        return False
    
    if s[l] in  '+-':
        l += 1
    if l > r:
        return False
    
    dot, ex = -1, -1
    for i in range(l, r + 1):
        if s[i] in '123456789':
            continue
        
        if s[i] == '.':
            if dot >= 0:
                return False
            dot = i
        elif s[i] in 'eE':
            if ex >= 0:
                return False
            ex = i
        else:
            return False
    
    if ex == l or ex == r:
        return False
    if dot == l:
        return dot < r and s[dot + 1] in '123456789'
    if dot == r:
        return dot > l and s[dot - 1] in '123456789'
    return True
        
```

```python
#int to roman:
def intToRoman(n):
    table = [['I', 'II', 'III', 'IV', 'V', 'VI', 'VII', 'VIII', 'IX'],
             ['X', 'XX', 'XXX', 'XL', 'L', 'LX', 'LXX', 'LXXX', 'XC'],
             ['C', 'CC', 'CCC', 'CD', 'D', 'DC', 'DCC', 'DCCC', 'CM'],
             ['M', 'MM', 'MMM']]

    re = ""
    strn = str(n)
    for i in range(len(strn)):
        if int(strn[i]) > 0:
            re += table[len(strn) - 1 - i][int(strn[i]) - 1]

    return re
print(intToRoman(99))
```

```python
#428 - power
def myPow(x, n):
    if x == 0:
        return 0
    if n == 0:
        return 1
    
    neg = True if n < 0 else False
    if neg:
        n = -n
    
    res = 1
    while n > 1:
        res *= x if n % 2 else 1
        x *= x
        n = n // 2
    
    res *= x
    if neg:
        res = 1 / res
    return res
print(myPow(2.0, 20))
```
