```python
class Solution:
    def evalRPN(self, tokens):
        st = []
        if tokens is None:
            return 0
        
        for t in tokens:
            if t is "+" or t is "-" or t is "*" or t is "/":
                if len(st) < 2:
                    return 0;
                
                r = st.pop(-1);
                l = st.pop(-1);
                
                if t is "+":
                    st.append(l + r)
                if t is "-":
                    st.append(l - r)
                if t is "*":
                    st.append(l * r)
                if t is "/" and r != 0:
                    st.append((int)(l / r))
            else:
                st.append((int)(t))
                
        if len(st) != 1:
            return 0;
        return st[-1];
    
```

```python
expression = ["(","(","1","+","(","2",")",")","*","3","+","(","4","/","2",")","-","5",")"]
def convertToPN(expression):
    if expression == []:
        return []
    
    ans = []
    st = []
    for exp in expression[::-1]:
        if exp == ")":
            st.append(exp)
        elif exp == "(":
            while st != [] and st[-1] != ")":
                ans.append(st.pop())
            
            if st != []:
                st.pop()
        elif exp == "+" or exp == "-":
            while st != [] and st[-1] != ")" and st[-1] != "+" and st[-1] != "-":
                ans.append(st.pop())
            st.append(exp)
        elif exp == "*" or exp == "/":
            st.append(exp)
        else:
            ans.append(exp)
    while st != []:
        ans.append(st.pop())
    return ans[::-1];
```

```python
def calculate(op, dt):
    r = dt.pop()
    l = dt.pop()
    if op == "+":
        return l + r
    elif op == "-":
        return l - r
    elif op == "*":
        return l * r
    elif op == "/" and r != 0:
        return (int)(l / r)
    else:
        return 0
        
def evaluateExpression(expression):
    if expression == []:
        return 0
    
    dt = []
    op = []
    for exp in expression:
        if exp == "(":
            op.append(exp)
        elif exp == ")":
            while op != [] and op[-1] != "(":
                if len(dt) < 2:
                    return 0
                dt.append(calculate(op.pop(), dt))
                
            if op != []:
                op.pop()
        elif exp == "+" or exp == "-":
            while op != [] and op[-1] != "(":
                if len(dt) < 2:
                    return 0
                dt.append(calculate(op.pop(), dt))
            op.append(exp)
        elif exp == "*" or exp == "/":
            while op != [] and op[-1] != "(" and op[-1] != "+" and op[-1] != "-":
                if len(dt) < 2:
                    return 0
                dt.append(calculate(op.pop(), dt))
            op.append(exp)
        else:
            dt.append((int)(exp))
    
    while op != []:
        if len(dt) < 2:
            return 0
        dt.append(calculate(op.pop(), dt))
        
    if len(dt) != 1:
        return 0
    return dt[-1]

print(evaluateExpression(["4", "-", "(", "2","-", "3", ")", "*", "2", "+", "5", "/", "5"]))
```

```python
class MyQueue:
    
    def __init__(self):
        self.st_pri = []
        self.st_sup = []


    def push(self, element):
        self.st_pri.append(element)

    def pop(self):
        if self.st_sup == [] and self.st_pri == []:
            return -1;
        elif self.st_sup == []:
            while self.st_pri != []:
                self.st_sup.append(self.st_pri.pop())
        
        return self.st_sup.pop()
    
    def top(self):
        if self.st_sup == [] and self.st_pri == []:
            return -1;
        elif self.st_sup == []:
            while self.st_pri != []:
                self.st_sup.append(self.st_pri.pop())
        
        return self.st_sup[-1]
```

```python
import random
class RandomizedSet:
    
    def __init__(self):
        self.dict = {}
        self.reg = []

    def insert(self, val):
        if val in self.dict:
            return
        
        self.dict[val] = len(self.reg)
        self.reg.append(val)
        
    def remove(self, val):
        if val not in self.dict:
            return
        
        idx = self.dict[val]

        self.reg[-1], self.reg[idx] = self.reg[idx], self.reg[-1]
        self.reg.pop()
        self.dict[self.reg[idx]] = idx
        del self.dict[val]

    def getRandom(self):
        if not self.reg:
            return -1
        
        idx = random.randint(0, len(self.reg) - 1)
        return self.reg[idx]

rset = RandomizedSet()
rset.insert(1)
print(rset.getRandom())
rset.insert(3)
rset.remove(1)
print(rset.getRandom())
```

```python
def mergekSortedArrays(arrays):
    ans,pq = [],[]
    sums = 0
    for i in range(len(arrays)):
        if not arrays[i]:
            continue
            
        sums += len(arrays[i])
        pq.append([arrays[i][0], i, 0])
    
    heapq.heapify(pq)
    while pq:
        val, row, col = pq[0][0], pq[0][1], pq[0][2]
        if col < len(arrays[row]) - 1:
            heapq.heappush(pq, [arrays[row][col + 1], row, col + 1])

        ans.append(heapq.heappop(pq)[0])

    return ans
```

```python
#124
def longestConsecutive(num):
    st = set()
    for v in num:
        if v not in st:
            st.add(v)
    
    ans = 1
    while st:
        cur = st.pop()
        left, right = cur - 1, cur + 1
        cnt = 1
        
        while st and left in st:
            st.remove(left)
            left -= 1
        
        while st and right in st:
            st.remove(right)
            right += 1
        
        ans = max(ans, right - left - 1)
    
    
    return ans
```

```python
#LC - contest
from collections import deque
import sys
class Solution(object):
    def maxSubarraySumCircular(self, A):
        presum = 0
        max_sum = max(A)
        
        length = len(A) 
        queue = deque([])
        for t in range(2):
            for i,v in enumerate(A):
                presum += v
                index = length * t + i
                while queue and queue[-1][0] > presum:
                    queue.pop()
                
                while queue and index - queue[0][1] > length:
                    queue.popleft()
                
                if queue:
                    max_sum = max(max_sum, presum - queue[0][0])
                
                queue.append((presum, index))
        
        return max_sum
sl = Solution()
print(sl.maxSubarraySumCircular([3,1,3,2,6]))
```
