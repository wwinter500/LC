````python
#414
import sys
def divide(dividend, divisor):
    #without using divide, multiply, mod
    #return int value
    INT_MAX = 2147483647 
    if divisor == 0:
        return INT_MAX
    
    neg = True if dividend ^ divisor < 0 else False
    a, b = abs(dividend), abs(divisor)
    ans, shift = 0, 31
    while shift >= 0:
        if a >= b <<shift:
            a -= b << shift
            ans += 1 << shift
        shift -= 1
    
    if neg:
        ans = -ans
    if ans > INT_MAX:
        return INT_MAX
    return ans
print(divide(20, 10))
````

````python
def searchMatrix(matrix, target):
    if not matrix or not matrix[0]:
        return 0
    
    n,m = len(matrix), len(matrix[0])
    i, j = n - 1, 0
    cnt = 0
    while i >= 0 and j < m:
        if matrix[i][j] == target:
            cnt += 1
            i -= 1
            j += 1
        elif matrix[i][j] < target:
            j += 1
        else:
            i -= 1
    return cnt
print(searchMatrix([[1, 3, 5, 7],[2, 4, 7, 8],[3, 5, 9, 10]], 3))
````

````python
#cat question- 1667
def binarysearchright(arr, target):
    l , r = 0 , len(arr) - 1
    while l < r:
        m = (r - l) // 2 + l
        if arr[m] >= target:
            r = m
        else:
            l = m + 1
    
    return l if arr[l] > target else len(arr)

def intervalStatistics(arr, k):
    onelist = []
    for i in range(len(arr)):
        if arr[i] == 1:
            onelist.append(i)

    res = 0
    n, n1 = len(arr), len(onelist)
    for i in range(n):
        if arr[i] == 1:
            continue

        j = binarysearchright(onelist, i)
        if j == len(onelist) or j + k >= len(onelist):
            #find how much zero after j
            zeros = n - i - (n1 - j)  
        else:
            zeros = n - i - k
            
        res += zeros + zeros * (zeros - 1) // 2
    
    return res
            
````

````python
#617 - maximum average subarray II : length at least k : maximum average
def maxAverage(nums, k):
    if not nums:
        return 0
    
    left,right = max(nums), min(nums)
    while right - left > 1e-5:
        mid = (left + right) / 2
        if presum_array(nums, k, mid):
            left = mid
        else:
            right = mid
    
    return left

def presum_array(nums, k, average):
    pre_sum = [0]
    for v in nums:
        pre_sum.append(pre_sum[-1] + v - average)
    
    pre_min = 0
    for i in range(k, len(nums) + 1):
        if pre_sum[i] - pre_min >= 0:
            return True
        
        pre_min = min(pre_min, pre_sum[i - k + 1])
    
    return False
````

````python
#search based on length
class Solution:
    def longestDupSubstring(self, S: str) -> str:
        #binary search length
        l, r = 1, len(S) - 1
        while l + 1 < r:
            m = (l + r) // 2
            cur = self.findDupSubstring(S, m)
            if cur != "":
                l = m
            else:
                r = m 
        
        return self.findDupSubstring(S, l)
    
    def findDupSubstring(self, s, length):
        sub = s[:length]
        subset = set([sub])
        for i in range(1,len(s) - length + 1):
            sub = sub[1:length] + s[i + length - 1]
            if sub not in subset:
                subset.add(sub)
            else:
                return sub
        #print(subset)
        return ""
````

````python
#437 - copy books
#divde to k partition, which has minimum maximun value
#solution with biarny search
def copyBooks(pages, k):
    if not pages:
        return 0
    
    start, end = max(pages), sum(pages) #theoritically minimum / maximum time
    while start + 1 < end:
        mid = (start + end) // 2  
        if get_least_people(pages, mid) <= k:
            end = mid
        else:
            start = mid

def get_least_people(pages, time_limit):
    count = 0
    time_cost = 0
    for page in pages:
        if time_cost + page > time_limit:
            count += 1
            time_cost = 0
            
        time_cost += page
    
    return count + 1
````

````python
#find a peak - 2d matrix
class Solution:
    def findPeakGrid(self, mat: List[List[int]]) -> List[int]:
        n, m = len(mat), len(mat[0])

        up = 0
        bottom = n - 1
        while up + 1 < bottom:
            mid = up + (bottom - up) // 2
            index = self.findMax(mat, mid)
            if mat[mid][index] < mat[mid - 1][index]:
                bottom = mid
            elif mat[mid][index] < mat[mid + 1][index]:
                up = mid
            else:
                return [mid, index]
        
        bottom_index = self.findMax(mat, bottom)
        up_index = self.findMax(mat, up)
        if mat[up][up_index] < mat[bottom][bottom_index]:
            return [bottom, bottom_index]
        else:
            return [up, up_index]
    
    def findMax(self, mat, row):
        maxv = max(mat[row])
        return mat[row].index(maxv)
````
