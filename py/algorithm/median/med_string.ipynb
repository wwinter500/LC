```python
#Repeated Substring Pattern
def getFactorList(n):
    factors = [1]
    nums = set([i for i in range(1, n // 2 + 1)])
    for f in range(2, n // 2 + 1):
        if f not in nums or n % f != 0:
            continue
        
        d = n // f
        nums.remove(f)
        factors.append(f)
        if d != f:
            nums.remove(d)
            factors.append(n // f)
    return factors
        
def subPattern(s):
    if len(s) == 1:
        return False
    
    n = len(s)
    factors = getFactorList(n)
    
    print(factors)
    for p in factors:
        sub = s[:p]
        start = p
        repeat = True
        while start < n:
            if sub != s[start : start + p]:
                repeat = False
                break
            start += p
        if repeat:
            return True
    return False

subPattern("aa")
```

```python
#Partition Labels
def partitionLabels(S):
    if not S:
        return []

    n = len(S)
    farpos = {}
    for i in range(n):
        farpos[S[i]] = i
    
    res = []
    l, r = 0, 0
    for i in range(n):
        if i > r:
            res.append(r - l + 1)
            l = i
        r = max(r, farpos[S[i]])

    if l < n:
        res.append(n - l)
    return res

#partitionLabels("ababcbacadefegdehijhklij")
partitionLabels("caedbdedda")
```

```python
#Number of Matching Subsequences
def numMatchingSubseq(self, s: str, words: List[str]) -> int:
    nxtPos = []
    n = len(s)
    tmp = [-1] * 26
    for i in range(n - 1, -1, -1):
        tmp[ord(s[i]) - ord('a')] = i
        nxtPos.append(tmp[:])
    nxtPos = nxtPos[::-1]

    ans = 0
    for word in words:
        if self.isSubseq(word, nxtPos):
            ans += 1
    return ans

def isSubseq(self, word, nxtPos):
    lenw, lens = len(word), len(nxtPos)
    i, j = 0, 0
    while i < lenw and j < lens:
        j = nxtPos[j][ord(word[i]) - ord('a')]
        if j < 0:
            return False
        i += 1
        j += 1
    return i == lenw
```
