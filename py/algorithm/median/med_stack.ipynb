```python
#mono stack
def trapwater(hist):
    if not hist or len(hist) < 2:
        return 0
    
    left, right = [hist[0]], [hist[-1]]
    maxv = max(hist)
    maxi = hist.index(maxv)
    
    res = 0
    for l in range(1, maxi):
        if hist[l] > left[-1]:
            left.append(hist[l])
        else:
            res += left[-1] - hist[l]

    for r in range(len(hist) - 2, maxi, -1):
        if hist[r] > right[-1]:
            right.append(hist[r])
        else:
            res += right[-1] - hist[r]
    return res

trapwater([0,1,0,2,1,0,1,3,2,1,2,1])
```

```python
#remove all occurrences of parts
def removeOccurrences(s, part):
    stack = []
    plen = len(part)
    print(plen)
    for ch in s:
        stack.append(ch)
        if len(stack) < plen:
            continue
            
        if ''.join(stack[-plen:]) == part:
            for _ in range(plen):
                stack.pop()
    
    return ''.join(stack)

removeOccurrences("daabcbaabcbc", "abc")
```

```python
#wiggle stack
def wiggleMaxLength(self, nums: List[int]) -> int:
    stack = []
    curr = 0
    for num in nums:
        if stack and stack[-1] == num:
            continue
        if len(stack) < 2:
            stack.append(num)
            if len(stack) == 2:
                curr = (stack[1] - stack[0]) // abs(stack[1] - stack[0])
        else:
            if (curr == -1 and num > stack[-1]) or (curr == 1 and num < stack[-1]):
                curr *= -1
            else:
                stack.pop()

            stack.append(num)

    return len(stack)
```

```python
class Solution:
    def evalRPN(self, tokens):
        st = []
        if tokens is None:
            return 0
        
        for t in tokens:
            if t is "+" or t is "-" or t is "*" or t is "/":
                if len(st) < 2:
                    return 0;
                
                r = st.pop(-1);
                l = st.pop(-1);
                
                if t is "+":
                    st.append(l + r)
                if t is "-":
                    st.append(l - r)
                if t is "*":
                    st.append(l * r)
                if t is "/" and r != 0:
                    st.append((int)(l / r))
            else:
                st.append((int)(t))
                
        if len(st) != 1:
            return 0;
        return st[-1];
```

```python
expression = ["(","(","1","+","(","2",")",")","*","3","+","(","4","/","2",")","-","5",")"]
def convertToPN(expression):
    if expression == []:
        return []
    
    ans = []
    st = []
    for exp in expression[::-1]:
        if exp == ")":
            st.append(exp)
        elif exp == "(":
            while st != [] and st[-1] != ")":
                ans.append(st.pop())
            
            if st != []:
                st.pop()
        elif exp == "+" or exp == "-":
            while st != [] and st[-1] != ")" and st[-1] != "+" and st[-1] != "-":
                ans.append(st.pop())
            st.append(exp)
        elif exp == "*" or exp == "/":
            st.append(exp)
        else:
            ans.append(exp)
    while st != []:
        ans.append(st.pop())
    return ans[::-1];
```

```python
def calculate(op, dt):
    r = dt.pop()
    l = dt.pop()
    if op == "+":
        return l + r
    elif op == "-":
        return l - r
    elif op == "*":
        return l * r
    elif op == "/" and r != 0:
        return (int)(l / r)
    else:
        return 0
        
def evaluateExpression(expression):
    if expression == []:
        return 0
    
    dt = []
    op = []
    for exp in expression:
        if exp == "(":
            op.append(exp)
        elif exp == ")":
            while op != [] and op[-1] != "(":
                if len(dt) < 2:
                    return 0
                dt.append(calculate(op.pop(), dt))
                
            if op != []:
                op.pop()
        elif exp == "+" or exp == "-":
            while op != [] and op[-1] != "(":
                if len(dt) < 2:
                    return 0
                dt.append(calculate(op.pop(), dt))
            op.append(exp)
        elif exp == "*" or exp == "/":
            while op != [] and op[-1] != "(" and op[-1] != "+" and op[-1] != "-":
                if len(dt) < 2:
                    return 0
                dt.append(calculate(op.pop(), dt))
            op.append(exp)
        else:
            dt.append((int)(exp))
    
    while op != []:
        if len(dt) < 2:
            return 0
        dt.append(calculate(op.pop(), dt))
        
    if len(dt) != 1:
        return 0
    return dt[-1]

print(evaluateExpression(["4", "-", "(", "2","-", "3", ")", "*", "2", "+", "5", "/", "5"]))
```

```python
import random
class RandomizedSet:
    
    def __init__(self):
        self.dict = {}
        self.reg = []

    def insert(self, val):
        if val in self.dict:
            return
        
        self.dict[val] = len(self.reg)
        self.reg.append(val)
        
    def remove(self, val):
        if val not in self.dict:
            return
        
        idx = self.dict[val]

        self.reg[-1], self.reg[idx] = self.reg[idx], self.reg[-1]
        self.reg.pop()
        self.dict[self.reg[idx]] = idx
        del self.dict[val]

    def getRandom(self):
        if not self.reg:
            return -1
        
        idx = random.randint(0, len(self.reg) - 1)
        return self.reg[idx]

rset = RandomizedSet()
rset.insert(1)
print(rset.getRandom())
rset.insert(3)
rset.remove(1)
print(rset.getRandom())
```

```python
#frequent stack - 895
#use array as stack element
#does not need insert back of freq - 1 after delete freq element
class FreqStack:
    def __init__(self):
        self.freq = {}
        self.cnts = []
        
    def push(self, x):
        if x not in self.freq:
            self.freq[x] = 0
        self.freq[x] += 1
        
        if self.freq[x] > len(self.cnts):
            self.cnts.append([])
        self.cnts[self.freq[x] - 1].append(x)
    
    def pop(self):
        if not self.stack or not self.stack[-1]:
            return -1
        
        x = self.stack[-1].pop()
        if not self.stack[-1]:
            self.stack.pop()
        
        self.freq[x] -= 1
        return x

            
```
