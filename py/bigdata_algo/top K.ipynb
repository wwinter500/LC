```python
#544 - 
def topk(nums, k):
    arr = nums[:k]
    heapq.heapify(arr)
    
    for i in range(k, len(nums)):
        heapq.heappush(arr, nums[i])
        heapq.heappop(arr)
    
    arr.sort()
    arr.reverse()
    return arr

#solution 2 - standard offline top k
def topk_standardoffline(nums, k):
    dt = {}
    for n in nums:
        if n not in dt.keys():
            dt[n] = 1
        else
            dt[n] += 1
    
    arr = []
    sums = 0
    for dk in dt.keys():
        sums += dt[dk]
        heapq.heappush(arr, [dk, dt[dk]])
        
        while arr[0][1] <= sums - k:
            sums -= arr[0][1]
            heapq.heappop(arr)
        
        arr[0][1] -= sums - k
    
    return arr
```

```python
#5. quick select
def kthLargestElement(self, n, nums):
    k = max(0, len(nums) - n)
    l, r = 0, len(nums) - 1

    return self.quickselect(nums, l, r, k)

def quickselect(self, nums, left, right, k):
    l,r = left, right
    m = l + (r - l) // 2
    pivot = nums[m]
    while l <= r:
        while l <= r and nums[l] < pivot:
            l += 1

        while l <= r and nums[r] > pivot:
            r -= 1

        if l <= r:
            nums[l], nums[r] = nums[r], nums[l]
            l += 1
            r -= 1

    if r >= k:
        return self.quickselect(nums, left, r, k)
    elif l <= k:
        return self.quickselect(nums, l, right, k)
    else:
        return nums[k]
```

```python
#545 - top k largest - online
class Solution:
    def __init__(self, k):
        self.k = k
        self.arr = []
    
    def add(self, num):
        heapq.heappush(self.arr, -1 * num)

    def topk(self):
        ans = []
        cnt = 0
        while self.arr and cnt < self.k:
            ans.append(-1 * heapq.heappop(self.arr))
            cnt += 1
            
        for val in ans:
            heapq.heappush(self.arr, -1 * val)
            
        ans.sort()
        ans.reverse()
        return ans
```
